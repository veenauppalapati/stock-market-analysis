Historical Stock Data Tool

Allows the user to visualize the stock price of a particular in a defined period of time.

```python
plt.figure(figsize=(10,6))
while(True):
    #Prompt the user to enter a company
    ticker = input("Enter a company's ticker (i.e. TSLA):")
    stock = yf.Ticker(ticker.upper())

    period_choice = input("Select a time period for stock data: 5d | 1mo | 3mo | ytd | 1y | 2y | 5y | self : ")
    if period_choice=="self":
        start_date = input("Enter a start date (yyyy-mm-dd): ")
        end_date = input("Enter an end date (yyyy-mm-dd): ")
        historical_data = stock.history(start = start_date, end = end_date)
    else:
        historical_data = stock.history(period=period_choice)


    plt.plot(historical_data['Close'])

    plt.title(f"Stock Prices of {ticker}")
    plt.grid(True)

    # Add the axis labels
    plt.xlabel('Date')
    plt.ylabel('Stock Price')

    # Display the graph with the added features
    plt.tight_layout()
    plt.draw()
    plt.show()

    repeat = input("Check another stock? Type \"Y\" to continue.")
    if(repeat.lower()!="y"):
        break;
```
